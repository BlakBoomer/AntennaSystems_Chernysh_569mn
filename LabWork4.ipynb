```python
#ОДНОСТРИЖНЕВА
#ПЛОЩИНА Н

import numpy as np
import matplotlib.pyplot as plt
import math

# Вхідні дані
F1h = []
FC = []
FB = []
FH = []
steps = []
SGP1 = 0
fS1 = 0
d_avg = 0.023   
l = 0.237   
wavelength = 0.043   
h = 0.10   
N = 2  
max_x_FB = []
max_y_FB = []
min_x_FB = []
min_y_FB = []
Zeros = []

# Однострижнева антена
ksi = 1 + wavelength / (2 * l)
print(f"ksi = {ksi:.3f}")

# Обчислення 
for teta in np.arange(0.01, np.pi / 2, 0.00001):
    mn1 = abs((ksi - 1) / (np.sin((np.pi * l / wavelength) * (ksi - 1))))
    mn2 = abs((np.sin((np.pi * l / wavelength) * (ksi - np.cos(teta)))) / (ksi - np.cos(teta)))
    mn3 = mn1 * mn2
    F1h.append(mn1)
    FC.append(mn2)
    FB.append(mn3)
    if 0.707 < mn3 < 0.708:
        SGP1 = 2 * math.degrees(teta)
        fS1 = mn3
    steps.append(math.degrees(teta))


cos_theta = np.cos(np.radians(steps))

max_x_FB = []
max_y_FB = []

for i in range(1, len(FB) - 1):
    if FB[i] > FB[i - 1] and FB[i] > FB[i + 1]:
        max_x_FB.append(steps[i])
        max_y_FB.append(FB[i])

for i in range(1, len(FB) - 1):
    if FB[i] < FB[i - 1] and FB[i] < FB[i + 1]:
        min_x_FB.append(steps[i])
        min_y_FB.append(0)
        
print("Ширина головної пелюстки в площині H = " + str(round(SGP1, 2)) + '\u00b0')

# Побудова графіків
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(steps, cos_theta, linewidth=0.7, label="$ \\cos(θ) $", linestyle='--')
ax.plot(steps, FB, linewidth=0.7, label="$ F_B(θ)=F_H(θ) $")


ax.plot(SGP1/2, fS1, 'ro', markersize=4, label="ШГП в площині H")
plt.annotate(f'({fS1:.3f}, {SGP1/2:.2f}\u00b0)',
                 xy=(SGP1 / 2, fS1),
                 xytext=((SGP1/2)+3, fS1+0.05),
                 arrowprops=dict(arrowstyle='->', color='black'), fontsize=6)
ax.plot([0, SGP1/2], [fS1, fS1], 'r--', linewidth=0.5)
ax.plot([SGP1/2, SGP1/2], [0, fS1], 'r--', linewidth=0.5)


plt.xticks(np.arange(0, 100, 2), fontsize=7)
plt.yticks(np.arange(0, 1.2, 0.1), fontsize=7)
plt.ylim(-0.01, 1.01)
plt.xlim(0, 90.5)

ax.plot(max_x_FB, max_y_FB, "o", markersize=4, color="black", label="$ \\theta_{max} $ H")
for x in max_x_FB:
    ax.vlines(x, 0, max_y_FB[max_x_FB.index(x)], color='gray', linestyle='--', linewidth=1)
ax.plot(min_x_FB, min_y_FB, "o", markersize=4, color="blue", label="$ \\theta_{min} $ H")

print("Табл. 1 - Значення нульових кутів")
print("------------------------------")
print("| № |  θ  |FB(θ)|")
m = 1
for i in min_x_FB:
    print(f"| {m} |{i:.2f}|  {0}  |")
    m += 1
print("------------------------------")


print("Табл. 2 - Значення максимальних кутів")
print("------------------------------")
print("| № |  θ  |FB(θ)|")
m = 1
for i in max_x_FB:
    print(f"| {m} |{i:.2f}| {max_y_FB[m-1]:.2f}|")
    m += 1
print("------------------------------")

# Налаштування графіка
ax.set_xlabel("θ (градуси)")
ax.set_ylabel("FE(θ),FH(θ)")
ax.set_title("Графіки  $ cos(θ) $ та $ F_H(θ) $")
ax.legend()
ax.grid()

plt.show()
```
